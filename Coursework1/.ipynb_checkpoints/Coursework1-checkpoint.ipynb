# Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
```

# Import assets

```python
mat_content = sio.loadmat('assets/face.mat')
```

# Functions

```python
def plot_face(mat_content, num):
    face_data = mat_content['X']
    face_label_data = mat_content['l']
    face_label = face_label_data[0, num]
    face = face_data[:, num]
    variant = num % 10 + 1
    plt.imshow(np.reshape(face,(46,56)).T, cmap = 'gist_gray')
    plt.title(f'Face {face_label} (Variant {variant})'), plt.xticks([]), plt.yticks([])
    
plt.subplot(211), plot_face(mat_content, 159)
plt.subplot(212), plot_face(mat_content, 151)

plt.show()
```
